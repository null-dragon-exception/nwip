```python
import pandas as pd
```

```python
localization_df = pd.read_csv('americas_localization - americas_localization.csv').fillna('')
localization_df
```

```python
default_column = 'default' #replace with 'modern' for AtE
full_loc = "l_english: #This file is autogenerated! Don't modify it directly!!!\n"
comment = 'comment_modern'

full_loc = full_loc+ '\n'.join(localization_df[localization_df[default_column].str.strip().str.len() > 0].apply(lambda x:f' {x.location}: "{x[default_column]}" #{x.modern}{", " if len(x[comment]) > 0 else ""}{x[comment]}', axis=1).values)

with open('nwip/main_menu/localization/english/location_names/nwip_location_names_l_english.yml', 'w', encoding='utf-8-sig') as file:
    file.write(full_loc)
```

```python
default_column = 'modern' #replace with 'modern' for AtE
full_loc = "l_english: #This file is autogenerated! Don't modify it directly!!!\n"
comment = 'comment_modern'

full_loc = full_loc+ '\n'.join(localization_df[localization_df[default_column].str.strip().str.len() > 0].apply(lambda x:f' {x.location}: "{x[default_column]}" #{x.default}{", " if len(x[comment]) > 0 else ""}{x[comment]}', axis=1).values)

with open('ate_exports/main_menu/localization/english/location_names/AUTOGEN_modern_location_names_l_english.yml', 'w', encoding='utf-8-sig') as file:
    file.write(full_loc)
```

```python
language_mapping = {'portuguese':'portuguese_language',
                   'spanish':'spanish_language',
                   'english':'english_language',
                    'french':'french_language',
                    'dutch':'dutch_dialect',
                    'russian':'russian_language',
                    'nahuatl':'nahuatl_language',
                    'mixtec':'mixtec_language',
                    'otomi':'otomi_language',
                    'german':'german_language',
                    'arawakan':'arawakan_language',
                   }
```

```python
total_new_locs = 0
for language in language_mapping:
    full_loc = "l_english: #This file is autogenerated! Don't modify it directly!!!\n"
    actual_language = language_mapping[language]
    column = language+'_hist'
    comment_column = 'comment_'+language+'_hist'
    
    full_loc = full_loc+ '\n'.join(localization_df[localization_df[column].str.strip().str.len() > 0].apply(lambda x:f' {x.location}.{actual_language}: "{x[column]}" #{x.default} ({x.modern}){", " if len(x[comment_column]) > 0 else ""}{x[comment_column]}', axis=1).values)
    total_new_locs +=len(localization_df[localization_df[column].str.strip().str.len() > 0].apply(lambda x:f' {x.location}.{actual_language}: "{x[column]}" #{x.default} ({x.modern}){", " if len(x[comment_column]) > 0 else ""}{x[comment_column]}', axis=1).values)
    with open(f'nwip/main_menu/localization/english/location_names/nwip_location_names_{language}_l_english.yml', 'w', encoding='utf-8-sig') as file:
        file.write(full_loc)
```
